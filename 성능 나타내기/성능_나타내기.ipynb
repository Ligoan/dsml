```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as LR
import matplotlib.pyplot as plt
import koreanize_matplotlib

df = pd.read_csv('/content/drive/MyDrive/데이터 과학과 머신러닝/population_2019.csv', encoding = 'cp949')
```

```python
x = np.array(range(0, 60)).reshape(-1, 1)
y = np.array(df['총계 여'][0:60]).reshape(-1, 1)
```

```python
model = LR(fit_intercept = True).fit(x, y)
```

```python
plt.title('선형회귀 모델의 예측과 실제 데이터 간 차이')
plt.xlabel('예측', )
plt.ylabel('잔차')
plt.scatter(model.predict(x), model.predict(x)- y)
plt.hlines(y = 0, xmin = model.predict(x).min(), xmax = model.predict(x).max(), color = 'red')
#plt.plot(x, model.coef_ * x + model.intercept_, 'b')
plt.show()
```

```python
x = np.array(range(61, 101)).reshape(-1, 1)
y = np.array(df['총계 여'][61:101]).reshape(-1, 1)
```

```python
model = LR(fit_intercept = True).fit(x, y)
```

```python
plt.title('선형회귀 모델의 예측과 실제 데이터 간 차이')
plt.xlabel('예측', )
plt.ylabel('잔차')
plt.scatter(model.predict(x), model.predict(x)- y)
plt.hlines(y = 0, xmin = model.predict(x).min(), xmax = model.predict(x).max(), color = 'red')
#plt.plot(x, model.coef_ * x + model.intercept_, 'b')
plt.show()
```

## 평균제곱오차(MSE)와 결정 계수(R^2)  

|구 분|내 용|
|---|---|
|**MSE**|각 데이터에 대한 잔차 제곱의 평균을 구한 것으로 일반적으로 값이 작을수록 성능이 뛰어남을 의미하는데, 이느 최소제곱법의 관점이 평가에 적용된 것이다.|
|**R^2**| 1-(MSE)(종속변수의 분산), 모델을 통해 설명할 수 있는 데이터의 비율을 나타내는 지표로 모델의 MSE와 종속변수의 분산 값을 이용하여 계산한다.
||0 이상 1 이하의 값으로 표현되며, 1에 가까울수록 모델의 성능으 뛰어남을 의미한다.|

```python
import pandas as pd
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('/content/drive/MyDrive/데이터 과학과 머신러닝/population_2019.csv', encoding = 'cp949')
```

```python
x = np.array(range(0, 60)).reshape(-1, 1)
y = np.array(df['총계 여'][0:60]).reshape(-1, 1)
```

```python
model = LR(fit_intercept = True).fit(x, y)
```

```python
print('MSE: ', mean_squared_error(y, model.predict(x)))
print('R^2: ', r2_score(y, model.predict(x)))
```

```python
x = np.array(range(61, 101)).reshape(-1, 1)
y = np.array(df['총계 여'][61:101]).reshape(-1, 1)
```

```python
model = LR(fit_intercept = True).fit(x, y)
```

```python
print('MSE: ', mean_squared_error(y, model.predict(x)))
print('R^2: ', r2_score(y, model.predict(x)))
```
